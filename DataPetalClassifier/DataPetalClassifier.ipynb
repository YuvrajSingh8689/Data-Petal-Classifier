```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("Iris.csv")
```

```python

# Display basic information about the dataset
print("Basic Information about the Dataset:")
print(df.info())
```

```python
# Display summary statistics of the dataset
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Display the number of samples for each species
print("\nNumber of Samples for Each Species:")
print(df['Species'].value_counts())
```

```python
# Box plot for each feature based on species
plt.figure(figsize=(15, 8))
plt.title("Box Plot for Each Feature Based on Species")
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.show()
```

```python
# Pairplot for visualizing relationships between variables
sns.pairplot(df, hue="Species", markers=["o", "s", "D"], palette="husl")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
plt.title("Correlation Heatmap")
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidths=.5)
plt.show()
```
